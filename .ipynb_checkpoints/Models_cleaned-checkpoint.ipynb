```python
C=10
delta_C=1
delta_P=0
time=10
alpha = 1.1
beta = 1.3
I_C = .7
P = 10
results = []
current_day = 0
delta_C = 1
rho_C =.6
rho_p = .3

while current_day <= time:
    
    # capital D represents delta
    
    dIdt = -rho_C * I_C
    
    dCdt = -rho_C*I_C
    C += dCdt
    
    dPdt = rho_C*I_C
    P += dPdt
    
    dDPdt = beta*rho_C*I_C
    
    dCdt = rho_C*I_C
    C += dCdt
    
    dPdt = -rho_C*I_C
    P += dPdt
    
    dEdt=0
    
    NI = rho_C * I_C
    
    dIdt = NI - rho_C
    
    dCdt = rho_C*I_C
    C += dCdt
    
    dDCdt = alpha * dCdt
    delta_C += dDCdt
    
    dDPdt = beta * dCdt
    delta_P += dDPdt
    
    rho_C = delta_C/C
    
    day_results = {
        'I_C': I_C,
        'C': C,
        'P': P,
        'delta_C': delta_C,
        'delta_P': delta_P,
    }
    
    results.append(day_results)
    current_day += 1
df = pd.DataFrame(results)
print(df)
```

## Somthing New
